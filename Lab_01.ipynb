```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[10,40,0,30,20,80]
y1=[40,10,50,30,20,70]
plt.plot(x,y,y1)
plt.xlabel("Year")
plt.ylabel("Growth Rate")
plt.title("Growth rate of Company")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,30,0,40,50]
z=[90,80,70,60,65]
j=plt.figure()
v=j.add_subplot(111,projection='3d')
v.plot(x,y,z)
v.set_xlabel("year")
v.set_ylabel("growth rate")
v.set_zlabel("Investment")
plt.show()
```

```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,30,0,40,50]
y1=[0,25,15,45,0]
y2=[40,20,10,5,0]
z=[90,80,70,60,65]
z1=[70,80,50,85,95]
z2=[75,45,25,5,65]
j=plt.figure()
v=j.add_subplot(111,projection='3d')
v.plot(x,y,z)
v.plot(x,y1,z1)
v.plot(x,y2,z2)
v.set_xlabel("year")
v.set_ylabel("growth rate")
v.set_zlabel("Investment")
v.set_title("investment of growth")
plt.show()
```

```python
import matplotlib.pyplot as plt
blood_groups = ['A', 'B', 'AB', 'O']
counts = [20, 70, 30, 50]
plt.bar(blood_groups, counts, color='purple')
plt.xlabel('Blood Groups')
plt.ylabel('Number of People')
plt.title('Distribution of Blood Groups')
plt.show()
```
