```python
import matplotlib.pyplot as plt
x=[1,2,3,4,5]
y=[10,30,0,40,50]
z=[90,80,70,60,65]
j=plt.figure()
v=j.add_subplot(111,projection='3d')
v.plot(x,y,z)
v.set_xlabel("year")
v.set_ylabel("growth rate")
v.set_zlabel("Investment")
plt.show()
```
